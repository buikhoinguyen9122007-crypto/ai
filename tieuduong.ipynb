```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
import ipywidgets as widgets
from IPython.display import display, clear_output
dulieugoc = pd.read_csv('diabetes.csv')
dactrung = dulieugoc.drop('Outcome', axis=1)
nhanbenh = dulieugoc['Outcome']
huatluayendactrung, kiemtradactrung, huanluyennhan, kiemtranhan = train_test_split(dactrung, nhanbenh, test_size=0.2, random_state=42)
bodinhchuan = StandardScaler()
huatluayendactrungchuanhoa = bodinhchuan.fit_transform(huatluayendactrung)
mohinhtho = Perceptron(max_iter=1000, eta0=0.1, random_state=42)
mohinhtho.fit(huatluayendactrung, huanluyennhan)
mohinhchuan = Perceptron(max_iter=1000, eta0=0.1, random_state=42)
mohinhchuan.fit(huatluayendactrungchuanhoa, huanluyennhan)
style_tieude = {'description_width': 'initial'}
layout_o_nhap = widgets.Layout(width='auto', margin='5px')
nhapmangthai = widgets.IntSlider(value=1, min=0, max=20, description='Số lần thai kỳ:', style=style_tieude, layout=layout_o_nhap)
nhapduonghuyet = widgets.IntSlider(value=100, min=0, max=250, description='Đường huyết:', style=style_tieude, layout=layout_o_nhap)
nhaphuyetap = widgets.IntSlider(value=70, min=0, max=150, description='Huyết áp:', style=style_tieude, layout=layout_o_nhap)
nhapdodayda = widgets.IntSlider(value=20, min=0, max=100, description='Độ dày da:', style=style_tieude, layout=layout_o_nhap)
nhapinsulin = widgets.IntSlider(value=80, min=0, max=900, step=5, description='Insulin:', style=style_tieude, layout=layout_o_nhap)
nhapbmi = widgets.FloatSlider(value=25.0, min=0.0, max=70.0, step=0.1, description='Chỉ số BMI:', style=style_tieude, layout=layout_o_nhap)
nhapphaghe = widgets.FloatSlider(value=0.5, min=0.0, max=3.0, step=0.01, description='Phả hệ:', style=style_tieude, layout=layout_o_nhap)
nhaptuoi = widgets.IntSlider(value=30, min=1, max=120, description='Tuổi tác:', style=style_tieude, layout=layout_o_nhap)
luachonthuattoan = widgets.ToggleButtons(options=[('Mô hình Chuẩn hóa', 'chuan'), ('Mô hình Thô', 'tho')], description='Chế độ:', button_style='info')
nutbamchandoan = widgets.Button(description='PHÂN TÍCH KẾT QUẢ', button_style='danger', layout=widgets.Layout(width='100%', height='50px'), icon='check')
khunghienthi = widgets.Output()
def thuchienchandoan(b):
    with khunghienthi:
        clear_output()
        dulieunguoidung = pd.DataFrame([{'Pregnancies': nhapmangthai.value, 'Glucose': nhapduonghuyet.value, 'BloodPressure': nhaphuyetap.value, 'SkinThickness': nhapdodayda.value, 'Insulin': nhapinsulin.value, 'BMI': nhapbmi.value, 'DiabetesPedigreeFunction': nhapphaghe.value, 'Age': nhaptuoi.value}])
        if luachonthuattoan.value == 'tho':
            ketquadudoan = mohinhtho.predict(dulieunguoidung)[0]
            dosaiso = accuracy_score(kiemtranhan, mohinhtho.predict(kiemtradactrung))
        else:
            dulieudaxuly = bodinhchuan.transform(dulieunguoidung)
            ketquadudoan = mohinhchuan.predict(dulieudaxuly)[0]
            dosaiso = accuracy_score(kiemtranhan, mohinhchuan.predict(bodinhchuan.transform(kiemtradactrung)))
        print(f"--- KẾT QUẢ PHÂN TÍCH (Độ tin cậy: {dosaiso * 100:.1f}%) ---")
        if ketquadudoan == 1:
            print("⚠️ CẢNH BÁO: Bạn có dấu hiệu nguy cơ mắc bệnh. Hãy đi khám bác sĩ.")
        else:
            print("✅ CHÚC MỪNG: Các chỉ số hiện tại nằm trong vùng an toàn.")
        print("-" * 50)
nutbamchandoan.on_click(thuchienchandoan)
tieudegiaodien = widgets.HTML("<h2>📊 CHẨN ĐOÁN TIỂU ĐƯỜNG SỚM</h2>")
luoithongso = widgets.GridBox([nhapmangthai, nhapduonghuyet, nhaphuyetap, nhapdodayda, nhapinsulin, nhapbmi, nhapphaghe, nhaptuoi], layout=widgets.Layout(grid_template_columns="repeat(2, 50%)"))
giaodienchinh = widgets.VBox([tieudegiaodien, luoithongso, widgets.HTML("<br>"), luachonthuattoan, nutbamchandoan, khunghienthi], layout=widgets.Layout(align_items='stretch', border='1px solid #ddd', padding='20px'))
display(giaodienchinh)
```
